```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import lxml
import json
from geopy.geocoders import Nominatim 
import requests 
from pandas.io.json import json_normalize 
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium 
```

#### Create soup object from the url

```python
page = urllib.request.urlopen("http://www.city-data.com/nbmaps/neigh-New-York-New-York.html")
soup = BeautifulSoup(page, 'html.parser')
```

#### Begin parsing for NYC Neighborhood Names

```python
neighborhoods = soup.find_all('span', class_='street-name')
neigh_array=[]
for neigh in neighborhoods:
    neigh_array.append(neigh.get_text())
```

#### Begin parsing for NYC Neighborhood Incomes

```python
neigh_data = soup.find_all("div", class_='hgraph')
data=[]
for i in neigh_data:
    if ('$' in (i.find("tr").get_text())):
        data.append(i.find("tr").get_text())
incomes=[]
not_listed=[]
for j in neigh_array:
    if any(j in string for string in data)==False:
        not_listed.append(j)
    for k in range(0,len(data)):
        if j in data[k]:
            incomes.append(data[k])
            break
```

```python
incomes2=[]
for x in incomes:
    incomes2.append(x.split(':')[1])
for y in not_listed:
    neigh_array.remove(y) 
```

#### Create Final Dataframe of the NYC Neighborhoods and Incomes

```python
df=pd.DataFrame(neigh_array,columns=['Neighborhood'])
df["Income"] = incomes2
df.head()
```

#### Create Geolocator object to locate the latitude/longitude of each neighborhood

```python
latitude_array=[]
longitude_array=[]
for location in df["Neighborhood"]:
    x=(location+ ', NY')
    geolocator = Nominatim(user_agent="ny_explorer")
    place = geolocator.geocode(x)
    if place is None:
        latitude_array.append(0)
        longitude_array.append(0)
    else:
        latitude_array.append((place.latitude))
        longitude_array.append((place.longitude))
```

```python
df['Latitude'] = latitude_array 
df['Longitude'] = longitude_array
df=df[df['Latitude']!=float(0)]
df.head()
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500,limit=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
nyc_venues = getNearbyVenues(names=df['Neighborhood'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                        )
```

```python
nyc_onehot = pd.get_dummies(nyc_venues[['Venue Category']], prefix="", prefix_sep="")
nyc_onehot['Neighborhood'] = nyc_venues['Neighborhood'] 
fixed_columns = [nyc_onehot.columns[-1]] + list(nyc_onehot.columns[:-1])
nyc_onehot = nyc_onehot[fixed_columns]
nyc_onehot.head()
```

#### Group each neighborhood by the number of existing gyms

```python
gym_counts = (nyc_gym.groupby('Neighborhood').count())['Venue']
gym_counts=gym_counts.reset_index()
```

```python
neighs=gym_counts["Neighborhood"].tolist()
counts=gym_counts["Venue"].tolist()
```

```python
total_gyms=[]
track=0
for name in df["Neighborhood"]:
    if name in neighs:
        total_gyms.append(counts[track])
        track=track+1
    else:
        total_gyms.append(0)
```
